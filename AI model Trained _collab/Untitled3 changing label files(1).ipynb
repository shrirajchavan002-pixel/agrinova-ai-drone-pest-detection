```python
import os

datasets = {
r"D:\YOLO_PROJECT":0,
r"C:\Users\anilk\OneDrive\Desktop\New folder\mildew.v1i.yolov8 (1)":1,
r"D:\FINAL YOLO PROJECT\early-blight.v2i.yolov8":2
}

for dataset_path, class_id in datasets.items():

    for split in ["train","valid"]:
        label_folder = os.path.join(dataset_path, split, "labels")

        if not os.path.exists(label_folder):
            continue

        for file in os.listdir(label_folder):

            file_path = os.path.join(label_folder,file)

            with open(file_path,"r") as f:
                lines = f.readlines()

            new_lines=[]

            for line in lines:
                parts=line.split()
                parts[0]=str(class_id)
                new_lines.append(" ".join(parts)+"\n")

            with open(file_path,"w") as f:
                f.writelines(new_lines)

print("Labels updated successfully")
```
