```python
import os

for path in ["/content/agri/agri/train/images",
             "/content/agri/agri/valid/images",
             "/content/agri/agri/train/labels",
             "/content/agri/agri/valid/labels"]:
    count = len(os.listdir(path))
    print(f"{'✅' if count > 0 else '❌'} {path.split('agri/agri/')[-1]}: {count} files")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

model.train(
    data="/content/agri/agri/dataset.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    name="agri_pest_v1",
    project="/content/runs",
    patience=20,
    augment=True,
    hsv_h=0.015,
    hsv_s=0.7,
    flipud=0.3,
    fliplr=0.5,
    save_period=10
)
```

```python
metrics = model.val()
print(f"mAP50:     {metrics.box.map50:.3f}")
print(f"mAP50-95:  {metrics.box.map:.3f}")
print(f"Precision: {metrics.box.mp:.3f}")
print(f"Recall:    {metrics.box.mr:.3f}")
```

```python
metrics_train = model.val(
    data="/content/agri/agri/dataset.yaml",
    split="train"   # ← this is the key change
)

print("=== TRAIN SET METRICS ===")
print(f"mAP50:     {metrics_train.box.map50:.3f}")
print(f"mAP50-95:  {metrics_train.box.map:.3f}")
print(f"Precision: {metrics_train.box.mp:.3f}")
print(f"Recall:    {metrics_train.box.mr:.3f}")
```
